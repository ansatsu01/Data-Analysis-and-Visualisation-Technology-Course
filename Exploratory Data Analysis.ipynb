# Let's find the number of films of the most widespread genres

```python
genres_dict = {'Drama' : films['Genre'].str.contains('Drama').sum(), 'War' : films['Genre'].str.contains('War').sum(),
           'Action' : films['Genre'].str.contains('Action').sum(), 'Romance' : films['Genre'].str.contains('Romance').sum(),
           'Crime' : films['Genre'].str.contains('Crime').sum(), 'Thiller' : films['Genre'].str.contains('Thriller').sum(),
           'Mystery' : films['Genre'].str.contains('Mystery').sum()}
genres = pd.Series(genres_dict)
```

```python
genres.plot.bar()
```

# The films of my favourite directors
